```python
import glob

path = r'C:\Users\neomo\Documents\GitHub\data_science_project_2\data' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

df = pd.concat(li, axis=0, ignore_index=True)
```

# Data Cleaning

```python
col_date=['state_changed_at','created_at','launched_at','deadline']
for i in col_date:
    df[i]=df[i].apply(lambda x:
    datetime.datetime.fromtimestamp(int(x)).strftime("%Y-%m-%d %H:%M:%S"))
```

```python
df.state_changed_at = pd.to_datetime(df.state_changed_at)
df.created_at = pd.to_datetime(df.created_at)
df.launched_at = pd.to_datetime(df.launched_at)
df.deadline = pd.to_datetime(df.deadline)
```

```python
df.drop(["permissions","is_backing","is_starred","friends","blurb","currency","currency_symbol", "currency_trailing_code",
        "current_currency","location","name","photo","source_url","urls","usd_type","profile","pledged","usd_pledged"], inplace=True, axis=1)
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df = df.sort_values(["id","state_changed_at"])
```

```python
df.drop_duplicates(subset="id", inplace=True, keep="last")
```

```python
df.drop(["id"], inplace=True, axis=1)
```

# Feature Engineering

## average pledge

```python
df["average_pledge"] = round((df.converted_pledged_amount / df.backers_count),2)
```

```python
df["average_pledge"].fillna(0, inplace=True)
```

## convert goal to USD

```python
df['usd_goal'] = round((df.goal * df.static_usd_rate), 2)
```

```python
df.drop(["goal","static_usd_rate","fx_rate"], axis=1, inplace=True)
```

## categories

```python
cat = [eval(i)["slug"] for i in df.category]
```

```python
cat = [elm.split("/") for elm in cat]
```

```python
main_cat = [elm[0] for elm in cat]
sub_cat = [elm[1] if len(elm) > 1 else elm[0] for elm in cat]
```

```python
df["main_cat"] = main_cat
df["sub_cat"] = sub_cat
df.drop("category", axis=1, inplace=True)
```

## profile ids

```python
ids = [int(re.findall(r'\d+', i)[0]) for i in df.creator]
```

```python
id_counts = [ids.count(elm) for elm in ids]
```

```python
df["profile_id_counts"] = id_counts
```

```python
df.profile_id_counts.value_counts()
```

```python
df.drop("creator", axis=1, inplace=True)
```

## slug length

```python
key_word_count = [len(elm.split("-")) for elm in df.slug]
```

```python
df["key_word_count"] = key_word_count
```

```python
df.drop("slug", axis=1, inplace=True)
```

```python
df.sub_cat.nunique()
```

## duration

```python
df['duration'] = df.deadline - df.launched_at
df['duration'] = df['duration'].dt.floor('d')
df['duration'] = df['duration'].dt.days.astype('int16')

df.groupby(['duration']).size()
```

```python
hist = df['duration'].hist(bins=10)
```

```python
df.drop(['state_changed_at','created_at','launched_at','deadline'], axis=1, inplace=True)
```
